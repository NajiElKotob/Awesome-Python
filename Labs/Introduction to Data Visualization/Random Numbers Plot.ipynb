```python
# Import matplotlib.pyplot
import matplotlib.pyplot as plt

# Import random
from random import randint as rnd
```

```python
d1 = [rnd(30, 50) for i in range(50)]
print(d1)
```

```python
plt.plot(d1)
plt.title('Random Numbers Plot')
plt.xlabel('Random Numbers between 30 and 50')
plt.show()
```
