### Bachelor's degrees by gender

```python
import pandas as pd
```

```python
dataset = 'https://raw.githubusercontent.com/NajiElKotob/Awesome-Python/master/Labs/Data/percent_bachelors_degrees_women_usa.csv'
```

```python
df = pd.read_csv(dataset)
print(df)
```

```python
print(df.columns)
# remove spaces in columns name
df.columns = (df.columns.str.replace(' ','_'))
df.columns = df.columns.str.lower()
print(df.columns)
```

```python
# Import matplotlib.pyplot
import matplotlib.pyplot as plt
```

```python
# Plot in blue the % of degrees awarded to women in the Physical Sciences
plt.plot(df.year, df.physical_sciences, 'r')

# Plot in red the % of degrees awarded to women in Computer Science
plt.plot(df.year, df.computer_science, 'b')
```

```python
# Display the plot
plt.show()
```
